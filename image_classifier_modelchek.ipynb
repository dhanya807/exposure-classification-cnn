```python
from google.colab import drive
drive.mount('/content/drive')

from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np

# Load the model from your Google Drive
model = load_model('/content/drive/MyDrive/exposure_classifier_model.h5')

# Path to a test image - **Change this to point to an actual image file**
# Replace 'your_image_file_name.jpg' with the actual name of an image file
img_path = '/content/drive/MyDrive/extracted_folder/For_Exposure/train_directory/poor contrast train/img13.jpeg'

# Load and preprocess the image
try:
    img = image.load_img(img_path, target_size=(128, 128))
    img_array = image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)  # Shape becomes (1, 128, 128, 3)

    # Predict
    predictions = model.predict(img_array)
    predicted_class = np.argmax(predictions)

    # Map class index to class name
    class_names = ['normal', 'poor contrast', 'underexposed', 'overexposed']  # Make sure this matches your folder names
    print("Predicted Class:", class_names[predicted_class])

except FileNotFoundError:
    print(f"Error: The image file was not found at {img_path}")
except IsADirectoryError:
    print(f"Error: The path {img_path} points to a directory, not an image file.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
